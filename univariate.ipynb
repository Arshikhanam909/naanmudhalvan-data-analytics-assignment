# **Univariate Analysis**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/content/archive (7) (1).zip')
print(data.head())
```

```python
sns.distplot(data.Date)
sns.displot(data.Lattitude)
```

```python
sns.histplot(data['Date'])
```

```python
sns.countplot(data['number of bathrooms'])
```

```python
plt.pie(data.Price)
plt.show()
```

```python
x = data['number of bedrooms'].value_counts()
plt.pie(x.values,
		labels=x.index,
		autopct='%1.1f%%')
plt.show()
```

# **Bivariate analysis**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
sns.barplot(x=data['id'], y=data['number of bedrooms'])
plt.xticks(rotation='vertical')
```

```python
sns.scatterplot(x=data['number of floors'],
				y=data['lot area'])
```

```python
sns.lineplot(x=data.Date,y=data.Lattitude)
```

```python
sns.scatterplot(x=data.Date,y=data.Lattitude)
```

# **Multivariate analysis**

```python
plt.boxplot(data)
```

```python
sns.jointplot(x='Built Year',y='Price',data=data)
```

```python
fig,ax = plt.subplots(2,2)
ax[0,0].plot(data.Date , data.Price)
ax[1,0].scatter(data.Date , data.Price)
ax[1,1].bar(data.Date , data.Price)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(data.corr(),annot=True)
```
